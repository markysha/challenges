```python
import pandas as pd
```

```python
import os
```

```python

df = {}

df["master"] = pd.DataFrame(columns=[
    "test",
    "algo version",
    "k",
    "time",
    "cut size",
])

df["updated"] = pd.DataFrame(columns=[
    "test",
    "algo version",
    "k",
    "time",
    "cut size"
])

with os.scandir('./') as entries:
    for entry in entries:
        if entry.name.split("_")[0] != "parhip":
            continue
        
        algo = entry.name.split("_")[1]
        test_num = entry.name.split("_")[2]
        k = entry.name.split("_")[6]
        configuration = entry.name.split("_")[-1]
        
        data = ""
        with open(entry.name) as f:
            data = f.read()
        
        lines = data.split("\n")
        
        n = lines[2].split(" ")[3]
        m = lines[2].split(" ")[5]
        time = lines[-4].split(" ")[-1]
        cut = lines[-3].split(" ")[-1]
        
        if configuration == "ecomesh":
            continue
            
#         print(algo, test_num, k, configuration, n, m, time, cut, data)
        df[algo] = df[algo].append({
            "test": test_num,
            "k": k,
            "algo version": algo,
            "time" : time,
            "cut size": cut
        },ignore_index=True)
```

```python
result = pd.merge(df["master"], df["updated"], on=["test", "k"], suffixes=[" master", " updated"])
```

```python
result = result.drop(columns = ["algo version master", "algo version updated"])
```

```python
result["test"] = pd.to_numeric(result["test"])
result["k"] = pd.to_numeric(result["k"])
result["time master"] = pd.to_numeric(result["time master"])
result["cut size master"] = pd.to_numeric(result["cut size master"])
result["time updated"] = pd.to_numeric(result["time updated"])
result["cut size updated"] = pd.to_numeric(result["cut size updated"])

result = result.sort_values(["test", "k"])
```

```python
result["time profit%"] = (result["time master"] - result["time updated"]) / result["time updated"] * 100
result["cut size profit%"] = (result["cut size master"] - result["cut size updated"]) / result["cut size updated"] * 100

result["time profit%"] = pd.Series(result["time profit%"], dtype=int)
result["cut size profit%"] = pd.Series(result["cut size profit%"], dtype=int)
# result["cut size profit%"] = pd.Series([int(round(time, 0)) for time in result["cut size profit%"]], dtype=int)
```
